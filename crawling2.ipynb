```python
#종합세트
import openpyxl
def write_excel (filename, sheetname, listdata):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width = 100
    excel_sheet.column_dimensions['B'].width = 20
    
    if sheetname != '':
        excel_sheet.title = sheetname
        
    for item in listdata:
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
    
    
```

```python
import requests
from bs4 import BeautifulSoup

product_list = []
for num in range(1,9):
    if num == 1:
        res = requests.get('https://davelee-fun.github.io/')
    else :
        res = requests.get('https://davelee-fun.github.io/page'+str(num))
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('div.card.h-100')
    for i in data:
        name = i.select_one('h4.card-text').get_text()
        explain = i.select_one('h2.card-title').get_text()
        data = i.select_one('span.post-date').get_text()
        #select_one 쓴 이유 = 바로 문자열을 추출하기 위해
        product_info = [name.strip(), explain.strip(), data.strip()]
        product_list.append(product_info)
write_excel('new file.xlsx','ProductList',product_list)
```

```python
index_list=list()
res = requests.get('https://finance.naver.com/sise/')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('ul.lst_major > li')
for i in data :
    name = i.select_one('a').get_text()
    number = i.select_one('span.up').get_text()
    move = i.select_one('em.bu_p.bu_pup').get_text()
    index_info = [name.strip(),number.strip(),move.strip()]
    index_list.append(index_info)
#엑셀 작성
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

for item in index_list : 
    print(item)
    excel_sheet.append(item)

excel_file.save('naver_finance.xlsx')
excel_file.close()
```
