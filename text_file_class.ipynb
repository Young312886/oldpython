```python
#copy 수작업
def copy(x,y):
    before = open(x,'r',encoding = 'UTF8')
    after = open (y, 'w', encoding = 'UTF8')
    before_lines = before.readlines()
    for lines in before_lines :
        after.write(lines)
    before.close()
    after.close()

copy ('text.txt', 'text3.txt')
```

```python
#암호화 하는 코드 
n = 0
new_word = []
def encoding (x):
    j = ord (x)
    k = j + 15
    k=chr (k)
    return k
def decoding (x):
    k = ord(x)
    j = k - 15
    j=chr(j)
    return j
## 또는 이 값을 if 로 변수를 정하는 방향으로 진행되어도 좋다
n = int(input('무슨 작업을 진행할 것 입니까?'))
intro = input('작업 할 파일명 : ')
outro = input ('저장할 파일명 :')

if n == 1:
    before = open(intro,'r',encoding = 'UTF8')
    after = open (outro, 'w', encoding = 'UTF8')
    list1 = before.readlines()
    for i in range(len(list1)):
        word = list1 [i]
        for j in word:
            new = encoding (j)
            new_word.append(new)
    Str=''.join(new_word)
    after.writelines(Str)
    before.close()
    after.close()
elif n == 2:
    before = open(intro,'r',encoding = 'UTF8')
    after = open (outro, 'w', encoding = 'UTF8')
    list1 = before.readlines()
    for i in range(len(list1)):
        word = list1 [i]
        for j in word:
            new = decoding (j)
            new_word.append(new)
    Str=''.join(new_word)
    after.writelines(Str)
    before.close()
    after.close()
else : 
    print ('잘못된 사항 입니다')
```

```python
import turtle
import random

#모양 - 펜 색상, 두께 랜덤
# 사각 = 모양의 하위 클래스, 모양 그리기

class Shape ():
    def __init__(self):
        self.kturtle = turtle.Turtle('turtle')
        
    def setPen(self):
        r = random.random()
        g = random.random()
        b = random.random()
        self.kturtle.pencolor((r,g,b))
        k = random.randrange(1,10)
        self.kturtle.pensize(k)
    def drawSahpe(self): ##없어도 문제는 없다
        pass
        
class Rectangle(Shape):
    def __init__(self, x, y):
        Shape.__init__(self)
        self.x = x
        self.y = y
        self.i = random.randrange(20,100)
        self.j = random.randrange(20,100)
    def drawShape(self):
        self.kturtle.penup()
        self.kturtle.goto(self.x+((self.i)/2),self.y-((self.i)/2))
        self.kturtle.pendown()
        self.setPen()
        for l in range(0,2):
            self.kturtle.right(90)
            self.kturtle.forward(self.i)
            self.kturtle.right(90)
            self.kturtle.forward(self.j)
            ##모든 변수에는 self를 붙여야 자신의 변수를 참조하게 된다. #그러니 무조건 self붙이기
            ##이전 클래스의 함수 불러오기는 간단하게 self.setPen()
            #더욱 깔끔하게 그릴려면 하나하나 좌표 설정 후, goto함수를 활용하면 깔끔하게 그려진다. 현재는 약간 이상한 위치에서 그려짐

def screenLeftClick(x,y):
    rect = Rectangle(x,y)
    rect.drawShape()
    
turtle.title("거북이로 직사각형 그리기")

turtle.onscreenclick(screenLeftClick,1)
turtle.done() ##반드시 괄호 붙이기 안그러면 loop 라는 오류 뜸
```

```python
import threading
import time

class RacingCar():
    carnumber = 0
    def __init__(self):
        RacingCar.carnumber += 1
    def racing (self):
        for i in range(0,3):
            print("%d 번째 자동차가 달리고 있습니다" %(RacingCar.carnumber))
            time.sleep(1)
            
car1 = RacingCar()
car2 = RacingCar()
car3 = RacingCar()

th1 = threading.Thread(target = car1.racing)
th2 = threading.Thread(target = car1.racing)
th3 = threading.Thread(target = car1.racing)

th1.start()
th2.start()
th3.start()
##스레드는 동시 진행이 가능. 작은 용량의 작업들을 동시 진행하는 방법
```
