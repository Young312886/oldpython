```python
#groupby() 즉, 선택한 항목에 따라 분류해 주고, 그를 분류기준으로 삼는다. 특히 앞으로 나옴
groupby_ = titanic.groupby('Pclass').mean()
groupby_
```

```python
titanic.groupby('Pclass')['Age'].agg([max,min,'mean'])#mean 은 함수가 없는듯하다
```

```python
#다양한 함수를 넣을수도 있다
format1 = {'Age':'max','SibSp':'sum','Fare':'mean'}
titanic.groupby('Pclass').agg(format1)
```

```python
import pandas as pd
titanic = pd.read_csv('titanic_train.csv')

titanic['Child_Adult'] = titanic.Age.apply(lambda x : 'Child' if x<= 15 else 'Adult') #apply (lambda x : 조건문)
titanic[['Age','Child_Adult']].head(10)
```

```python

titanic['Age_cat'] = titanic.Age.apply(lambda x : 'Child' if x <= 15 else('Adult' if x <=60 else 'Elderly') )

titanic.Age_cat.value_counts()
```
