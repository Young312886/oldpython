```python
import re
import openpyxl

regex = re.compile(' [A-Za-z].')#이게 정규표현식

work_book = openpyxl.load_workbook('train.xlsx')

work_sheet = work_book.active
for each_row in work_sheet.rows:
    print(each_row[3].value)
    print(regex.findall(each_row[3].value))

work_book.close()
#이렇게 같은 규칙을 가지고 있으면 호칭을 정해서 모든 단어에서 필요한 부분을 따로 
#추출 할 수 있다.
```

```python
import re
string = "(DAVE)"
re.sub('[^A-Z]', '',string) # 앞의 조건에 해당하는 문자를 두번째 문자로 대체해라#그래서 문자가 아닌 ()가 ""(공백)으로 대체됨
#참고로 re.sub는 변수가 아니기에 = 을 한번 쓰면 원하는 값이 나오지 않는다. 주의하자!!!
```

점은 줄바꿈을 제외하고 모든 문자를 지칭할 수 있다.
즉 'D.A' 는 앞뒤가 D,A인 모든 문자들이다.(사이에 아무거나 들어가도 됨)
.은 하나의 문자만
?는 앞의 문자가 0,1번(둘다) 반복되는 패턴
*는 앞의 문자 0번 반복 (0개 또는 이상, 즉, 없어도 무관)
+는 앞의 문자 1번 반복 (1번 이상. 1번은 있어야)
단, 둘다 동일한 문자여야 한다는 것이다.(바로 앞의 문자)

```python
pattern = re.compile('D.A')
pattern.search("DAA") #맞으면 이러한 결과값이 도출
```

```python
#점을 넣고 싶으면 \. 을 넣으시오 [.]이거도 됨
string = "DDA D1A DDA DA DDDA DageA"

```

```python
import re
re.sub('D.A','Dave',string)
```

{n} 앞 문자가 n번 반복

```python
pattern = re.compile('AD{2,5}A')
```

```python
pattern.search('ADDA')
```

```python
pattern.search('ADDDDDDA')#6번이라 해당 안됨
```

```python
import re
pattern = re.compile('[a-zA-Z]+')#알파벳 전체가 반복되는 문자열이 있는가
matched = pattern.search("DAve123") #안에 패턴에 부합하는 문자열이 존재하는가?
a_matched = pattern.match("DDD2") #이 문자열 시작부터 패턴에 부합하는 문자열이 있는가?
print(matched)
print(a_matched)
```

```python
#findall 표현식과 매칭되는 모든 문자열 리턴
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Game of the throne and the Fire')
print(findalled) 
```

```python
#findall을 활용한 판단문
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Game of the throne and the Fire')
if len(findalled)>0 :
    print('존재')
else :
    print("없다")
```

```python
pattern2 = re.compile(':') #스플릿. 정규표현식을 기준으로 쉽게 문자열 가르기 할 수 있다.
splited = pattern2.split('python:thorne')네
print(splited)
```

```python
pattern2 = re.compile(" [A-Z]{2} ")
splited = pattern2.split('asdf VS asdflkj')
print(splited)
```

```python
#sub를 활용한 문자열 대체
pattern2 = re.compile(":")
subed = pattern2.sub('**','asdoi :asdfoij : asf')
print(subed)
```

```python
#또는 바로 sub를 활용하여 문자열 대체
subed = re.sub(":","**","asdfo : asdoif : ads")
print(subed)
```

```python
pattern = re.compile('-[0-9]{7}')
subed = pattern.sub("-*******","970408-1067185")
print(subed)
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import pprint


res = requests.get('http://corners.gmarket.co.kr/Bestsellers')
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.select('div.best-list')
bestlist = data[1]
data1 = bestlist.select('ul>li')

#문자열 측정을 통해 오류 방지
link_re = re.compile('http://')

#액셀
excel_file = openpyxl.Workbook()
excel_sheet=excel_file.active
product_info = list()
column_data = ['순서','상품명','상품가격','회사이름','상품링크']
excel_sheet.append(column_data)
for index,value in enumerate(data1):
    name = value.select_one('a.itemname').get_text()
    price = value.select_one('span').get_text()
    base= value.select_one('div.thumb > a')
    link = base.get('href')
    if link_re.match(link): #문자열 검사 이후 작동되게 만듦
        res = requests.get(link)
        soup = BeautifulSoup(res.content, 'html.parser')
        company = soup.select_one('p > span.text__seller > a').get_text()    
    #print(index+1,'\b.상품명:',name,'\n상품가격 :', price, '\n상품 링크:',link, '\n회사 이름:',company) #b는 앞의 빈칸을 없애는 백스페이스
    
    product_info = [index +1,name,price,company,link]
    excel_sheet.append(product_info)
    #if index == 9:
        #break 로 짧게 볼 수 있다.
excel_file.save('Gmarket.xlsx')
excel_file.close()




    
```
