데이터 프레임에서 ndarray로 = 그냥 단순히 dataframe.values 하면 자체로 ndarray가 형성 (type) = ndarray, shape도 확인 가능

리스트로 변환하려면, dataframe.values.tolist() 하면 리스트 형태로 변환 (ndarray랑은 약간 다른 형태로 출력된다

딕셔너리로 변환하려면, dataframe.to_dict('list') 하면 딕셔너리로. key는 column 이름들로 되어 있다. 'list'빼면 약간 다른 형태로 바뀐 데이터가 나온다. (딕셔너리 안에 딕셔너리)


```python
titanic['Age_0']=0
titanic.head(3)
```

```python
titanic['Age_by_10'] = titanic.Age*10
titanic.head(3)
```

```python
titanic['Family_no']=titanic.SibSp+titanic['Parch']+1
titanic.head(4)
```

```python
#데이터 삭제
#axis = 0 이면 index, axis = 1이면 column 삭제
#본문을 수정하고 싶으면inplace = True 
drop1 = titanic.drop('Age_0',axis = 1)
titanic.head(3) #inplace 가 false면 기존의 데이터에서는 삭제가 되지 않음 (단순히 drop1은 삭제된 모습이 저장된다.)
#반대로 true 면 기존의 데이터가 수정, output은 없다
```

```python
drop = titanic.drop('Age_by_10', axis = 1)
titanic.head(3)
```

```python
titanic_drop = titanic.drop(['Family_no','Age_by_10','Age_0'], axis =1, inplace = True) #한번에 삭제 반환되는 값은 none
titanic.head(3)
```

```python
titanic_drop = titanic.drop([0,1], axis = 0)
print(titanic_drop) 
```

```python
indexes = titanic.index
print(indexes)
print(type(indexes)) #이 친구는 index 라는 value 이다
print(indexes.values) #numpy니가 .values를 붙여야 callable
```

```python
print(indexes.values.shape)
print(indexes[:5].values) #.values 안적으면 range 형태로만 출력된다 values의 앞뒤는 약간의 의미 차이가 생긴다
#숫자 하나일 경우에는 print(indexes[5])가능은 함
# 참고로 수정은 불가능(상호 작용이 안된다)
```

```python
Fare = titanic['Fare']

print('fair max = ', Fare.max())
print('fair sum = ', Fare.sum())
print('fair sum` =', sum(Fare)) # 자리수가 다르다. 설정이 다르기 때문
print('합 =', (Fare +3).head(3))
```

```python
#reset_index => 기존의 인덱스들 대신 새로 인덱스 부여. (여기도 inplace 설정해 줄 수 있다) 기존의 건 그대로 들어가짐
titanic_new = titanic.reset_index(inplace = True)
print(titanic_new)
print(titanic.head(3))
```

```python
drop = titanic.drop('index', inplace= True, axis = 1)
print(titanic.head(3)) #만약 level_0가 생겼으면, 이는 reset이 두번 진행된 것이다. 수작업으로 없애 줘라
```
