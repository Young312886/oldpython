```python
import pandas as pd
import numpy as np
import statsmodels
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
mydata = pd.read_csv('AirPassengers.csv')
mydata
```

```python
mydata.rename(columns = {'#Passengers':'y','Month':'ds'},inplace = True)
mydata.ds = pd.to_datetime(mydata['ds'])
mydata
```

```python
model = Prophet(seasonality_mode='additive')
model.fit(mydata)
future = model.make_future_dataframe(periods = 12*3, freq='MS')
forecast = model.predict(future)
model.plot(forecast)
plt.tight_layout()
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(mydata)
```

```python
future = model.make_future_dataframe(periods = 365*3)
forecast = model.predict(future)
model.plot(forecast)
```

```python
future = model.make_future_dataframe(periods = 12*3, freq = 'MS')
forecast = model.predict(future)
model.plot(forecast)
```

```python
#@title
model.plot_components(forecast)
```

```python
# divvy 자료 분석
data = pd.read_csv('/content/divvy_hourly.csv')
data
```

```python
df = pd.DataFrame({'ds':pd.to_datetime(data.date),'y':data.rides})
df
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df)
future = model.make_future_dataframe(periods = 365*24,freq='h')
forecast = model.predict(future)
```

```python
model.plot(forecast)
```

```python
model.plot_components(forecast)
```

```python
df_new = df[(df['ds'].dt.hour >= 8)&(df['ds'].dt.hour<18)]
df_new
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(df_new)
future = model.make_future_dataframe(periods = 365*24,freq='h')
forecast = model.predict(future)
```

```python
model.plot(forecast)
```

```python
fig = model.plot(forecast)
plt.xlim(pd.to_datetime(['2018-08-01','2018-08-04']))
plt.ylim(-2000,4000)
plt.show() #3일 데이터만 관측
```

```python
future2 = future[(future['ds'].dt.hour >= 8) & (future['ds'].dt.hour < 18)]
forecast2 = model.predict(future2)
fig = model.plot(forecast2)
plt.show()
```
