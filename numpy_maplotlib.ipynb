```python
score = np.array([30,50,46,25,83,42,21,68,92])
b = score > 40
b = b * score
for i in range (len(score)):
    if b[i] ==0:
        b[i] = b[i]
    else :
        b [i] = b[i]+10

#원했던거 score + b *10

print(b)
        
```
